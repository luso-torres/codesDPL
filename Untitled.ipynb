# MVDR Power to LCMV

```python
import numpy as np
import matplotlib.pyplot as plt

# Problem setup
lambda_w = 1  # Normalize wavelength to 1
k = 2 * np.pi / lambda_w  # Wave number

# Transmitter array (31 elements, λ/2 spacing)
N_tx = 31
d_tx = np.linspace(-(N_tx-1)/2, (N_tx-1)/2, N_tx) * (lambda_w / 2)

# Receiver array (11 elements, 2λ spacing)
N_rx = 11
d_rx = np.linspace(-(N_rx-1)/2, (N_rx-1)/2, N_rx) * (2 * lambda_w)

# Scattering points
scat_points = [(40 * lambda_w, -30 * lambda_w), (70 * lambda_w, 50 * lambda_w)]

# Function to compute steering vectors
def steering_vector(d_array, theta):
    return np.exp(-1j * k * d_array * np.cos(theta))

# Angles for power calculation
angles = np.linspace(-np.pi/2, np.pi/2, 180)  # From -90° to 90°

# Compute received signal covariance matrix R
R = np.zeros((N_rx, N_rx), dtype=complex)

for x_s, y_s in scat_points:
    theta_s = np.arctan2(y_s, 100 * lambda_w)  # Angle of arrival
    a_rx = steering_vector(d_rx, theta_s)
    R += np.outer(a_rx, a_rx.conj())  # Contribution from each scatterer

R /= len(scat_points)  # Normalize

# Define constraint matrix C (assume unit gain in two scattering directions)
C = np.column_stack([steering_vector(d_rx, np.arctan2(y, 100*lambda_w)) for _, y in scat_points])
d = np.ones((C.shape[1], 1))  # Desired response vector

# Compute LCMV weight matrix A
R_inv = np.linalg.pinv(R)  # Inverse of R
A = R_inv @ C @ np.linalg.inv(C.T @ R_inv @ C) @ d

# Compute MVDR power for each angle
P_mvdr = np.array([1 / np.abs(steering_vector(d_rx, theta).conj().T @ R_inv @ steering_vector(d_rx, theta))
                   for theta in angles])

# Normalize power
P_mvdr_dB = 10 * np.log10(P_mvdr / np.max(P_mvdr))

# Plot power pattern
plt.figure(figsize=(8, 5))
plt.plot(np.degrees(angles), P_mvdr_dB, label="MVDR Power")
plt.axvline(np.degrees(np.arctan2(-30, 100)), linestyle="--", color="r", label="Scattering Point 1")
plt.axvline(np.degrees(np.arctan2(50, 100)), linestyle="--", color="g", label="Scattering Point 2")
plt.xlabel("Angle (degrees)")
plt.ylabel("Power (dB)")
plt.title("MVDR Power Response")
plt.legend()
plt.grid()
plt.show()

# Return estimated A
print("Estimated A using LCMV:\n", A)
```

# Convex Minimization



```python
import numpy as np
import scipy.linalg as linalg
import matplotlib.pyplot as plt

# Define the number of elements and the wavelength
N = 31  # Number of elements
lambda_ = 1  # Wavelength (normalized)
d = lambda_ / 2  # Inter-element spacing (lambda/2)
k = 2 * np.pi / lambda_  # Wavenumber

# Define the desired directions (angles in degrees) for MIMO system
theta_desired = [20, 40]  # Example: Two desired directions for MIMO
alpha = 45  # Angular range for sidelobe constraint

# Function to compute steering vector for a given angle
def steering_vector(theta, N, d, k):
    theta_rad = np.radians(theta)  # Convert to radians
    n = np.arange(N)  # Element indices
    a = np.exp(1j * k * d * n * np.sin(theta_rad))  # Steering vector
    return a

# Compute the steering vectors for the desired directions
a_d = [steering_vector(theta, N, d, k) for theta in theta_desired]  # Desired direction steering vectors

# Define the covariance matrix (for simplicity, assume identity matrix as a placeholder)
R = np.eye(N)  # Covariance matrix of the received signal

# Define the constraints: desired directions
# Desired direction: a_d^H * w = 1 for each desired direction

# Constraints matrix (stacked for MIMO)
A = np.vstack([ad.conj() for ad in a_d])  # Stack the desired constraints (each row is a steering vector)
b = np.array([1] * len(a_d))  # Right-hand side of constraints for desired directions

# Function to check if the sidelobe constraint is satisfied
def sidelobe_constraint(theta_scan, w, threshold_dB=-40, alpha=45):
    power_pattern = np.zeros_like(theta_scan)
    for i, theta in enumerate(theta_scan):
        a = steering_vector(theta, N, d, k)  # Compute steering vector for each scan angle
        power_pattern[i] = np.abs(np.dot(w.conj(), a))**2  # Calculate the power

    # Normalize the power pattern
    power_pattern_db = 10 * np.log10(power_pattern / np.max(power_pattern))

    # Apply the sidelobe constraint: Power outside [-alpha, alpha] must be below threshold_dB
    outside_range = np.abs(theta_scan) > alpha
    sidelobe_violations = power_pattern_db[outside_range] > threshold_dB
    return np.sum(sidelobe_violations)  # Return number of violations (want to minimize this)

# Solve the LCMV optimization problem using a linear system
w = linalg.lstsq(A.T @ R @ A, A.T @ R @ b)[0]  # Solve using least squares

# Normalize the weights (optional)
w = w / np.linalg.norm(w)

# Define the scan angles
theta_scan = np.linspace(-90, 90, 500)  # Scan angles

# Check the sidelobe constraint
violations = sidelobe_constraint(theta_scan, w)
print(f"Number of sidelobe violations: {violations}")

# Plot the power pattern (beamforming radiation pattern)
power_pattern = np.zeros_like(theta_scan)
for i, theta in enumerate(theta_scan):
    a = steering_vector(theta, N, d, k)  # Compute steering vector for each scan angle
    power_pattern[i] = np.abs(np.dot(w.conj(), a))**2  # Calculate the power

# Normalize the power pattern and convert to dB
power_pattern_db = 10 * np.log10(power_pattern / np.max(power_pattern))

# Plot the power pattern
plt.figure(figsize=(8, 6))
plt.plot(theta_scan, power_pattern_db)
plt.title('Beamforming Power Pattern (LCMV) with Sidelobe Constraint')
plt.xlabel('Scan Angle (Degrees)')
plt.ylabel('Power (dB)')
plt.grid(True)
plt.xlim([-90, 90])
plt.ylim([-50, 0])

# Mark the desired directions
for theta in theta_desired:
    plt.axvline(x=theta, color='r', linestyle='--', label=f'Desired Direction {theta}°')

# Mark the angular range for sidelobe constraint
plt.axvspan(-alpha, alpha, color='gray', alpha=0.2, label=r'$\pm\alpha$ Range')

plt.legend()
plt.show()
```

# Full algorithm

```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt

# Constants
lambda_wavelength = 1  # Normalize wavelength to 1
d_tx = lambda_wavelength / 2  # TX element spacing
d_rx = 2 * lambda_wavelength  # RX element spacing
num_tx = 31  # TX elements
num_rx = 11  # RX elements
l = 100 * lambda_wavelength  # Distance between TX and RX
alpha = np.radians(45)  # Define alpha (10 degrees example)

# Scattering points
scatterers = np.array([[40, -30], [70, 50]]) * lambda_wavelength

# Element positions
tx_positions = np.linspace(-(num_tx - 1) / 2, (num_tx - 1) / 2, num_tx) * d_tx
rx_positions = np.linspace(-(num_rx - 1) / 2, (num_rx - 1) / 2, num_rx) * d_rx

# Steering vector function for each element and angle
def steering_vector(positions, theta):
    """Computes the steering vector for given positions and angles."""
    return np.exp(1j * 2 * np.pi * positions[:, None] * np.sin(theta) / lambda_wavelength)

# Angles for power calculation
angles = np.linspace(-np.pi / 2, np.pi / 2, 180)
angles_outside_main_lobe = np.concatenate([
    np.linspace(-np.pi / 2, -alpha, 50),
    np.linspace(alpha, np.pi / 2, 50)
])

# Compute steering vectors for both TX and RX
steering_tx = steering_vector(tx_positions, angles)
steering_rx = steering_vector(rx_positions, angles)

# Initial A estimate (without optimization)
A = np.random.randn(num_tx, num_rx) + 1j * np.random.randn(num_tx, num_rx)

# Compute Power before optimization
P_before_opt = np.array([
    np.abs(np.conj(steering_tx[:, i]).T @ A)**2 for i in range(len(angles))
])

# Normalize to make the maximum gain 0 dB
P_before_opt_max = np.max(P_before_opt)
P_before_opt_normalized = P_before_opt / P_before_opt_max  # Normalize

# Convex Optimization to optimize A
A_var = cp.Variable((num_tx, num_rx), complex=True)

# Compute TX power response outside main lobe
P_sidelobes = [
    cp.norm(np.conj(steering_tx[:, i]).T @ A_var, 'fro')**2
    for i in range(len(angles_outside_main_lobe))
]

# Constraints: sidelobes < -40dB
constraints = [P_sidelobes[i] <= 10**(-40 / 10) for i in range(len(P_sidelobes))]

# Objective: Minimize norm of A (to preserve power pattern)
objective = cp.Minimize(cp.norm(A_var, 'fro'))

# Solve optimization problem
prob = cp.Problem(objective, constraints)
prob.solve()

# Extract optimized A
A_optimized = A_var.value

# Compute Power after optimization
P_after_opt = np.array([
    np.abs(np.conj(steering_tx[:, i]).T @ A_optimized)**2 for i in range(len(angles))
])

# Normalize to make the maximum gain 0 dB after optimization
P_after_opt_max = np.max(P_after_opt)
P_after_opt_normalized = P_after_opt / P_after_opt_max  # Normalize

# Convert power to dB after normalization
P_before_opt_dB = 10 * np.log10(P_before_opt_normalized)
P_after_opt_dB = 10 * np.log10(P_after_opt_normalized)

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(np.degrees(angles), P_before_opt_dB, label="Before Optimization", linestyle="dashed")
#plt.plot(np.degrees(angles), P_after_opt_dB, label="After Optimization", color='r')
plt.xlabel("Angle (degrees)")
plt.ylabel("Power (dB)")
plt.title("MVDR Power Response Before and After Optimization")
plt.legend()
plt.grid()
plt.show()
```
